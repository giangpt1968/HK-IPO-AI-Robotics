### 2.1 KPI helpers

These functions standardize common IPO metrics:
- Growth (YoY), annualized run-rate (for interim periods)
- Gross margin (if gross profit available)
- Net margin


````python
def yoy_growth(curr, prev):
    if prev is None or pd.isna(prev) or prev == 0 or pd.isna(curr):
        return np.nan
    return (curr / prev) - 1.0

def safe_div(a, b):
    if b is None or pd.isna(b) or b == 0 or pd.isna(a):
        return np.nan
    return a / b

def annualize_interim(value, months=6):
    # Simple annualization: value * (12 / months)
    if value is None or pd.isna(value):
        return np.nan
    return value * (12 / months)

kpi = financials.copy()
# FY2023: no prior year; FY2024: YoY vs FY2023; 1H2024: interim (no YoY); 1H2025: HoH vs 1H2024
kpi["revenue_yoy"] = [
    np.nan,
    yoy_growth(kpi.loc[1, "revenue"], kpi.loc[0, "revenue"]),
    np.nan,
    yoy_growth(kpi.loc[3, "revenue"], kpi.loc[2, "revenue"]),
]
kpi["gross_margin"] = kpi.apply(lambda r: safe_div(r["gross_profit"], r["revenue"]), axis=1)
kpi["net_margin"] = kpi.apply(lambda r: safe_div(r["net_income"], r["revenue"]), axis=1)
kpi
````

### 5.4 Cash Burn & Runway Analysis

**Critical for loss-making AI/Robotics IPOs** - when do they run out of money?

````python
# --- Cash Burn & Runway Analysis ---
# Extract from prospectus cash flow statement

# Pre-IPO cash position (from balance sheet)
cash_pre_ipo_rmb = 94_457e3  # As at Jun 30, 2025 (update from prospectus)

# Net IPO proceeds
net_proceeds_hkd = deal["net_proceeds_hkd"]
hkd_to_rmb = 0.92  # Approximate exchange rate
net_proceeds_rmb = net_proceeds_hkd * hkd_to_rmb

# Total cash post-IPO
total_cash_post_ipo_rmb = cash_pre_ipo_rmb + net_proceeds_rmb

# Estimate quarterly cash burn (from operating cash flow in prospectus)
# Use most recent interim period as proxy
quarterly_operating_cf_rmb = -89_087e3 / 2  # 1H2025 net loss / 2 quarters (rough proxy)

# Runway calculation
if quarterly_operating_cf_rmb < 0:
    runway_quarters = total_cash_post_ipo_rmb / abs(quarterly_operating_cf_rmb)
    runway_years = runway_quarters / 4
else:
    runway_quarters = np.inf
    runway_years = np.inf

cash_runway = {
    "cash_pre_ipo_rmb_m": cash_pre_ipo_rmb / 1e6,
    "net_proceeds_rmb_m": net_proceeds_rmb / 1e6,
    "total_cash_post_ipo_rmb_m": total_cash_post_ipo_rmb / 1e6,
    "quarterly_burn_rmb_m": abs(quarterly_operating_cf_rmb) / 1e6,
    "runway_quarters": round(runway_quarters, 1),
    "runway_years": round(runway_years, 1),
    "runway_status": "OK (>24m)" if runway_years > 2 else "CAUTION (<24m)" if runway_years > 1 else "CRITICAL (<12m)"
}

pd.Series(cash_runway).to_frame("Value")
````
