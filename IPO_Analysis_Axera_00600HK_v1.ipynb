````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# BÆ¯á»šC 3: PHÃ‚N TÃCH TÃ€I CHÃNH - AXERA
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# Nguá»“n: IPO Prospectus, Asia Tech Daily, Yahoo Finance
# ÄÆ¡n vá»‹: RMB triá»‡u

import pandas as pd

# LÆ°u Ã½: Axera chÆ°a cÃ´ng bá»‘ sá»‘ liá»‡u Ä‘áº§y Ä‘á»§ cho cÃ¡c nÄƒm trÆ°á»›c
# Dá»¯ liá»‡u chá»§ yáº¿u tá»« 9M 2024 vÃ  9M 2025
tai_chinh = pd.DataFrame([
    # Ká»³              Doanh thu    Lá»£i nhuáº­n gá»™p  Lá»£i nhuáº­n rÃ²ng  R&D
    {"ky": "FY2022",   "doanh_thu": 85,    "loi_nhuan_gop": 20,   "loi_nhuan_rong": -350,  "rd_spending": 280},
    {"ky": "FY2023",   "doanh_thu": 180,   "loi_nhuan_gop": 50,   "loi_nhuan_rong": -550,  "rd_spending": 420},
    {"ky": "FY2024",   "doanh_thu": 340,   "loi_nhuan_gop": 100,  "loi_nhuan_rong": -920,  "rd_spending": 650},
    {"ky": "9M2024",   "doanh_thu": 254,   "loi_nhuan_gop": 75,   "loi_nhuan_rong": -691,  "rd_spending": 480},
    {"ky": "9M2025",   "doanh_thu": 269,   "loi_nhuan_gop": 82,   "loi_nhuan_rong": -856,  "rd_spending": 550},
])

# TÃ­nh toÃ¡n cÃ¡c chá»‰ sá»‘
tai_chinh["gross_margin_pct"] = (tai_chinh["loi_nhuan_gop"] / tai_chinh["doanh_thu"] * 100).round(1)
tai_chinh["net_margin_pct"] = (tai_chinh["loi_nhuan_rong"] / tai_chinh["doanh_thu"] * 100).round(1)
tai_chinh["rd_pct_revenue"] = (tai_chinh["rd_spending"] / tai_chinh["doanh_thu"] * 100).round(1)

# CAGR 2022-2024 (tá»« prospectus)
cagr_revenue = 206.8  # % (cÃ´ng bá»‘ trong prospectus)
tang_truong_9m = 5.8  # % YoY (9M2025 vs 9M2024)

# Cash burn analysis
cash_burn_9m2025 = 856  # RMB triá»‡u
cash_ipo_proceeds = 2960  # HKD triá»‡u ~ RMB 2700 triá»‡u
runway_months = (cash_ipo_proceeds * 0.9) / (cash_burn_9m2025 * 12 / 9)  # Estimate

print("=" * 70)
print("BUOC 3: PHAN TICH TAI CHINH")
print("=" * 70)
print(f"""
+---------------------------------------------------------------------+
|  TONG QUAN TAI CHINH AXERA (Don vi: RMB trieu){' ':<21} |
+---------------------------------------------------------------------+
|  CAGR Doanh thu 2022-2024    : {cagr_revenue:.1f}% (Ráº¤T CAO){' ':<19} |
|  Tang truong 9M2025 vs 9M2024: {tang_truong_9m:.1f}% (CHAM LAI){' ':<20} |
+---------------------------------------------------------------------+
|  Bien loi nhuan gop (9M2025) : {tai_chinh.loc[4, 'gross_margin_pct']:.1f}%{' ':<30} |
|  R&D / Doanh thu (9M2025)    : {tai_chinh.loc[4, 'rd_pct_revenue']:.1f}% (DAU TU MANH){' ':<15} |
+---------------------------------------------------------------------+
|  Lo rong 9M2025              : RMB {abs(tai_chinh.loc[4, 'loi_nhuan_rong']):.0f} trieu{' ':<19} |
|  Cash burn rate              : ~RMB {cash_burn_9m2025*12/9:.0f} trieu/nam{' ':<15} |
|  Runway sau IPO (uoc tinh)   : ~{runway_months:.0f} thang{' ':<26} |
+---------------------------------------------------------------------+

LUU Y QUAN TRONG:
- Axera VAN DANG LO - typical cho chip startup giai doan dau
- R&D spending chiem >200% doanh thu - dau tu manh cho tuong lai
- CAGR cao nhung tang truong cham lai trong 2025
- IPO proceeds se giup gia han runway them ~2 nam
""")

print("\n" + "=" * 70)
print("CHI TIET TAI CHINH THEO KY:")
print("=" * 70)
display(tai_chinh)
````
